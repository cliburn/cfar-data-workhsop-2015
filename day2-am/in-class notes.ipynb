```python
hiv_data.head(10)
```

```python
cols_to_groupby = ['participant_id', 'analyte']
cols_to_view = ['fi-bkgd-neg', 'fi-bkgd']
f_means = hiv_data.groupby(cols_to_groupby).mean()[cols_to_view]
f_means
```

```python
f_means.unstack()
```

```python
f_means.unstack(level=0)
```

```python
f_means.unstack(level=1)
```

```python
plt.figure(figsize=(12, 8))

sns.boxplot(x='analyte', y='fi-bkgd-neg', hue='participant_id',
           data=hiv_data)
```

```python
plt.figure(figsize=(12, 8))

sns.boxplot(y='analyte', x='fi-bkgd-neg', hue='participant_id',
           data=hiv_data)
```

```python
hiv_data.set_index(['visit_date', 'participant_id'])
```

```python
cols_to_groupby = ['visit_date', 'participant_id', 'analyte']
data_grp = hiv_data.groupby(cols_to_groupby)
```

```python
data_grp_means = data_grp.mean()
data_grp_means
```

Now, pull fi-bkgd-neg column out of the group means dataframe:

```python
f_grp_mean = data_grp_means['fi-bkgd-neg']
f_grp_mean
```

```python
analyte_df = f_grp_mean.unstack(level=2)
analyte_df
```

```python
hiv_data_time = analyte_df.unstack(level=1)
hiv_data_time
```

```python
p24_time_data = hiv_data_time.p24
p24_time_data
```

```python
p24_time_data.interpolate().plot()
```
